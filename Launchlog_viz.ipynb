```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
__author__ = 'Marcel-Jan Krijgsman'

launch_file_name = "https://www.planet4589.org/space/log/launchlog.txt"

# Positions of columns
colspecs=[[0,8], [13,24], [25,32], [40,51], [55,85], [86,111], [112,118], [121,143],
                  [144,159], [160,192], [193,194], [198,211]]

# Names of the columns
colnames=['launch', 'launch_date', 'launch_time_utc', 'cospar_nr', 'payload_name', 'orig_payload_name',
          'satcat_nr', 'launchveh_type', 'launch_veh_serie', 'launch_site', 'outcome', 'reference']
```

```python
df = pd.read_fwf(launch_file_name, header=1, colspecs=colspecs, names=colnames)
```

```python
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
# Convert the launch_date to a date format
df["launch_date"] = pd.to_datetime(df["launch_date"])
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
# The original data has empty fields on the same launch. This fills them with the same launch details.
df.launch = df.launch.fillna(method='ffill')
df.launch_date = df.launch_date.ffill()
df.launch_time_utc = df.launch_time_utc.ffill()
df.launchveh_type = df.launchveh_type.ffill()
df.launch_veh_serie = df.launch_veh_serie.ffill()
df.launch_site = df.launch_site.ffill()
df.outcome = df.outcome.ffill()
df.reference = df.reference.ffill()
```

```python
df.tail()
```

```python
# Only successful launches
dfs = df[df['outcome'] == 'S']
```

```python
# Successful launches since 2015
dfs15 = dfs[dfs['launch_date'] > '2015-01-01']
```

```python
dfs15.head()
```

```python
dfs15g = dfs15.groupby(pd.Grouper(key="launch_date", freq="M")).count()
```

```python
dfsg = dfs.groupby(pd.Grouper(key="launch_date", freq="M")).count()
```

```python
dfs15g.head()
```

```python
dfs15gl = pd.DataFrame(dfs15g, columns =['launch'])
```

```python
dfsgl = pd.DataFrame(dfsg, columns =['launch'])
```
